#### Sr.no., MovieName, year

```python
def movies():
    for i in driver.find_elements('xpath', '//h3[@class="lister-item-header"]'):
        x=i.text.split(". ")
    
        sno.append(x[0])    # s.no.
    
'''        if len(x)>2:                     #movie
            y=('. ').join(x[1:])
            movie.append(y.split(' (')[0])
        else:
            movie.append(x[1].split(' (')[0])
        
        year.append(i.text.split(' (')[1][:-1])    #year'''
```

#### Duration, Certification, Genre

```python
def genre():
    for i in driver.find_elements('xpath', '//p[@class="text-muted "]'):
        if len(i.text.split(' | '))>2:
            cert.append(i.text.split(' | ')[0])    #certification
            dur.append(i.text.split(' | ')[1])    # duration
            gen.append(i.text.split(' | ')[2].split(', '))    #genre
            
        else:
            cert.append(np.nan)    #certification
            dur.append(i.text.split(' | ')[0])    # duration
            gen.append(i.text.split(' | ')[1].split(', '))    #genre
```

#### rate, metascore

```python
def metascore():
    for i in driver.find_elements('xpath', '//div[@class="ratings-bar"]'):
        rate.append(i.text[:3])    #rating
    
        try:
            meta.append(i.text.split("\n")[1][:2])    #metascore
        except:
            meta.append(np.nan)
```

#### Director Name, stars

```python
def star():
    for i in driver.find_elements('xpath', '//p[@class=""]'):
        x=i.text.split(" | ")
    
    
        #Directors
        y= x[0].split(": ")
        z=y[1].split(', ')
        
        direct1.append(z[0])
        try:
            direct2.append(z[1])
        except:
            direct2.append(np.nan)
        
        try:
            direct3.append(z[2])
        except:
            direct3.append(np.nan)
        
        
        #stars
        y= x[1].split(": ")[1].split(", ")
    
        star1.append(y[0])
        try:
            star2.append(y[1])
        except:
            star2.append(np.nan)
        
        try:
            star3.append(y[2])
        except:
            star3.append(np.nan)
        
        try:
            star4.append(y[3])
        except:
            star4.append(np.nan)
```

#### votes, Gross collection

```python
def votes():
    for i in driver.find_elements('xpath', '//p[@class="sort-num_votes-visible"]'):
        x=i.text.split(" | ")
    
        y= x[0].split(": ")
        vote.append(y[1])    #votes
    
    
        #gross collection
        try:
            y= x[1].split(": ")
            gross.append(y[1][1:])
        except:
            gross.append(np.nan)
```

#### Popularity

```python
def pop():
    for i in range (len(link)):
        driver.get(link[i])
        time.sleep(1)
        print(i, end="\t")
        try:
            popul.append(driver.find_element('xpath', '//div[@class="sc-5f7fb5b4-1 bhuIgW"]').text)
            
        except:
            popul.append(np.nan)
            
        movie.append(driver.find_element('xpath', '//span [@class="sc-afe43def-1 fDTGTb"]').text)    #movie name
        
        #year
        year.append(driver.find_element('xpath', '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[1]/a').text)
```

#### Next button

```python
def nxt():
    try:
        driver.find_element('xpath', '//a[@class="lister-page-next next-page"]').click()
            
    except:
        print("No next button found/ List ended.")
        pass
```

#### Data loading

```python
link=[]
sno, movie, year=[],[],[]
cert, dur, gen=[],[],[]
rate, meta=[],[]
direct1=[]
direct2=[]
direct3=[]
star1, star2, star3, star4= [],[],[],[]
gross=[]
vote=[]
popul=[]

c=1

while True:
    print("Scrapping page no {0}".format(c))
    for i in range (1,51):
        try:
            l=driver.find_element('xpath', '//*[@id="main"]/div/div[3]/div/div['+ str(i)+ ']/div[3]/h3/a')
            link.append(l.get_attribute("href"))
        except:
            print(len(link), "no. of data is scrapped")
            break
        
        
    movies()
    genre()
    metascore()
    star()
    votes()
    nxt()
    time.sleep(3)
    
    if len(link)%50 != 0:
        break
        
    c+=1
```
